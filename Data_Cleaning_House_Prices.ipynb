### House Prices Data sets

```python
import pandas as pd

df = pd.read_csv("datasets/HousePrices.csv")
print(df.head())
```

```python
# Data Cleaning

print("Data Information : \n")
print(f"{df.info()}\n")

print(f"Data Shape : \n{df.shape}")

# Remove nulls
df.dropna(inplace=True)

# Remove invalid values
df = df[df['price'] > 0]
df = df[df['bedrooms'] > 0]
df = df[df['bathrooms'] > 0]

# Fix renovated year
df.loc[df['yr_renovated'] == 0, 'yr_renovated'] = df['yr_built']

# Convert date
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year


# Remove outliers
df = df[df['price'] < df['price'].quantile(0.99)]

# Reset index
df.reset_index(drop=True, inplace=True)

print("Data cleaned successfully ✅")
print(f"{df.info()}\n")
df.isnull().sum()
df.describe()




    
```

```python
# Mean median ,Std etc of selected data.

cols = ["sqft_lot","yr_built","price"]

for col in cols:
    print(f"Mean : {df[col].mean()}")
    print(f"Median : {df[col].median()}")
    print(f"Standard Deviation : {df[col].std()}")
```

```python
# Correlation

print(f"Sqft Lot and price Correlation :{df["sqft_lot"].corr(df["price"])}")
print(f"Year Built and price correlation : {df["yr_built"].corr(df["price"])}")
```
